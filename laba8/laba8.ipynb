Вам необходимо провести базовый EDA выбранного набора данных.

Требования к анализу:

построить не менее 4 визуализаций различных видов;
каждая визуализация должным образом оформлена и читается даже в отрыве от контекста;
по каждой визуализации необходимо написать вывод (какую гипотезу на ее основе можно выдвинуть?).

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(
    '2020.csv')
df = data[['Country name', 'Ladder score']].head(10)
df = df.groupby(['Country name']).max()
df.plot(kind='bar', title = 'Топ 10 самых счастливых стран')
```

Следовательно, самая счастливая страна - Финляндия, а наименее счастливая - Люксембург (из топ 10и)

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(
    '2020.csv')
df = data[['Regional indicator', 'Logged GDP per capita']]
df = df.groupby(['Regional indicator']).mean().sort_values(by='Logged GDP per capita', ascending=False)
df.plot(kind='pie', y='Logged GDP per capita', title = 'Топ 10 ВВП по регионам')
```

Видим, что топ 1 регион по ВВП - Северная Амера, а на 10ом месте - Африка (к Югу от Сахары)

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(
    '2020.csv')
df = data[['Regional indicator', 'Social support']]
df = df.groupby(['Regional indicator']).mean().sort_values(by='Social support', ascending=False).head(10)
df.plot(kind ='barh')
```

Видим, что рейтинг соц поддержки наиболее высокий у региона Северной Америки, а наиболее низкий - у Африки (южнее Сахары)

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(
    '2020.csv')
data = data[['Country name', 'Generosity']]
data = data.groupby(['Country name']).max().sort_values(by='Generosity', ascending=False).head(5)
data.plot(title = 'Топ 5 самых щедрых стран')
```

По данному графику видим, что лидирует Мьянма, а отстает от всех Кения
